```python
import graphviz

# Crear el diagrama de flujo
dot = graphviz.Digraph(
    "SpecificGeneticAlgorithmSteps",
    comment="Detailed Flowchart for the TFM Genetic Algorithm",
    format="png",
)

# Configuración global de nodos y bordes
dot.attr("node", shape="box", style="filled,rounded", color="lightblue", fontname="Helvetica")
dot.attr("edge", fontname="Helvetica")

# --- Definición de Nodos ---

# Inicio
dot.node("start", "START", shape="ellipse", color="#3498db", fontcolor="white")

# Paso 1: Definición de Codificación y Parámetros
dot.node(
    "define",
    "DEFINE ENCODING & PARAMETERS",
)

# Paso 2: Inicialización
dot.node(
    "initialization",
    "INITIALIZATION",
)

# Paso 3: Evaluación de la Aptitud
dot.node(
    "evaluate",
    "EVALUATE FITNESS\n\n",
)

# Decisión de Terminación (Diamante)
dot.node(
    "dec_termination",
    "TERMINATION DECISION",
    shape="diamond",
    color="#f1c40f",
)

# Fin
dot.node(
    "end", "END", shape="ellipse", color="#2c3e50", fontcolor="white"
)

# Paso 4: Aplicar Elitismo
dot.node(
    "elitism",
    "APPLY ELITISM",
)

# Paso 5: Selección
dot.node(
    "selection",
    "SELECTION",
)

# Paso 6: Cruce
dot.node(
    "crossover",
    "CROSSOVER",
)

# Paso 7: Mutación
dot.node(
    "mutation",
    "MUTATION",
)

# Paso 8: Reproducción
dot.node(
    "reproduction",
    "REPRODUCTION",
)

# --- Definición de Bordes (Flechas) ---

# Flujo Inicial
dot.edge("start", "define")
dot.edge("define", "initialization")
dot.edge("initialization", "evaluate")

# Flujo Principal y Bucle
dot.edge("evaluate", "dec_termination")

# Decisión: Sí (Terminar)
dot.edge("dec_termination", "end", label="YES")

# Decisión: No (Continuar Bucle)
dot.edge("dec_termination", "elitism", label="NO")
dot.edge("elitism", "selection")
dot.edge("selection", "crossover")
dot.edge("crossover", "mutation")
dot.edge("mutation", "reproduction")

# Flecha de Vuelta a la Evaluación
dot.edge("reproduction", "evaluate", label="NEW POPULATION")

# Generar y mostrar la imagen
dot.render("GeneticAlgorithmFlowchart", view=True)

print("Diagrama de flujo generado correctamente como 'GeneticAlgorithmFlowchart.png'.")
```
